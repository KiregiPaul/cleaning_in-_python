```python
df.rename(columns= {'IMBD title ID': 'IMBD_title_ID', 'Original titlÊ' : 'Original_title', 'Release year' : 'Release_year',
                    'Genrë¨' : 'Genre', 'Duration' : 'Duration', 'Country' : 'Country', 'Content Rating' : 'Content_Rating',
                    'Director' : 'Director', 'Unnamed: 8': 'null', 'Income' : 'Income', ' Votes ' : 'Votes', 'Score' : 'Score'}, inplace=True)
```

```python
df.drop(columns= {'null'}, inplace = True) #dropping unwanted ccolumns
```

```python
df.columns # columns after dropping the null column
```

```python
df.fillna({'Duration':'Unknown'}, inplace = True)
```

```python
df.fillna({'Content_Rating':'Not Rated'}, inplace = True)
# handling missing values
```

```python
# dropping rows with critical missing data
df.dropna(subset= ['IMBD_title_ID', 'Original_title', 'Release_year'], inplace=True)
```

```python
# converting data types
df['Release_year'] = pd.to_datetime(df['Release_year'], errors='coerce')
df['Income'] = pd.to_numeric(df['Income'].replace('[$,]', '', regex=True), errors='coerce')
df['Votes'] = pd.to_numeric(df['Votes'].replace('[.,]', '', regex=True), errors='coerce')
```

```python
df
```

```python
# clean and standardize the data the 'score' column
df['Score'] = pd.to_numeric(df['Score'].replace('^0-9.]', '', regex=True), errors='coerce')
```

```python
# normalize any scores greater than 10
df['Score'] = df['Score'].apply(lambda x: x/10 if x > 10 else x)
```
